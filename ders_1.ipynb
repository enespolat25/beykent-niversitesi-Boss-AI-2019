Modeli oluşturun meyvenin Elma mı yoksa Portakal mı olduğunu sınıflandırın.

**Veri Kümesini Oluştur:** 

Veri Kümesi, özelliklerin ve etiketlerin koleksiyonundan başka bir şey değildir. Meyvelerin sınıflandırılması için gerekli olan girdilerdir. Etiketler, sorunun çıktılarıdır (Bizim sorunumuzda Elma ve Portakal etiketlerdir)

Not: Veri Kümesi daha doğru tahminler daha doğru.

Sorunumuz için veri kümesi:

```python
from IPython.display import Image
Image(filename='resimler/elma.PNG')
```

Örneğin, meyve kabuğu düz ve 130 g ağırlığında ise, Elma olarak sınıflandırılır. Aynı şekilde, Meyve pürüzlü ve yaklaşık 170 g ağırlığında ise Portakal olarak sınıflandırılır.

Şimdi veri setini Python koduna dönüştürebiliriz.

```python
ozellikler = [[130, 'duz'], 
              [140, 'duz'], 
              [150, 'puruzlu'], 
              [170, 'puruzlu']] 
etiketler = ['elma', 'elma ', 'portakal', 'portakal']
```

Şimdi yukarıda metinsel geçen özellikleri rakamsal dönüşümle ifade deceğiz.

```python
ozellikler = [[130,1], 
              [140,1], 
              [150,0], 
              [170, 0]]
etiketler = [0,0,1,1]
```

Özellik olarak, '1'  düz kabukluyu '0 pürüzlü kabukluyu temsil eder. Etiketlerde '0' elmayı, '1' portakali temsil eder.

**Modeli Oluştur**

Karar Ağaçları (DT'ler)  bir makine öğrenme yöntemidir. Amaç, veri özelliklerinden çıkarılan basit karar kurallarını öğrenerek bir hedef değişkenin değerini öngören bir model oluşturmaktır. Basit Sözlerle Karar ağacı, verilen girdiyi temel alarak kararı verir.

```python
Image(filename='resimler/treee.PNG')
```

Şimdi Modelimizi oluşturabiliriz.

Modeli oluşturmadan önce, Sklearn'ün ne olduğunu bilmeliyiz.

Sklearn, birçok Sınıflandırma, regresyon ve kümeleme algoritmasına sahip olan Açık kaynaklı makine öğrenme kütüphanesidir. Sorunumuzu çözmek için sklearn'den DecisionTreeClassifier algoritmasını kullanabiliriz.

```python
from sklearn import tree

ozellikler = [[130,1],[140,1], [150,0], [170, 0]]
etiketler = [0,0,1,1]
siniflandirici = tree.DecisionTreeClassifier()
```

Bu noktada, modeli yeni yaptık. Ancak , verilen meyvenin elma mı yoksa portakal mı olduğunu tahmin edemiyoruz. Modelimiz meyveyi tahmin etmek zorundaysa, modeli Özellikler ve Etiketler ile eğitmek zorundayız. Karşılık gelen girdiler için karşılık gelen çıktıyı haritalamak gerekir). Başka bir deyişle, meyve pürüzsüz ve 130 g ağırlığında ise, 'Elma'  olması daha muhtemeldir.

**Modeli Eğit**

fit komutu ile modelimizi eğitelim

```python
model = siniflandirici.fit (ozellikler , etiketler)
```

**Tahmin Yap**

predict komutu ile tahminimizi modele yaptırabiliriz.

```python
tahmin=model.predict([[170,0]])
tahmin=int(tahmin)
```

```python
etiket={0:"elma",1:"portakal"}
print(etiket[tahmin])
```

# Faydalı olması dileğiyle

```python
tahmin=model.predict([[2000,0]])
tahmin=int(tahmin)
```

```python
etiket={0:"elma",1:"portakal"}
print(etiket[tahmin])
```
